2. How does broadcasting work in NumPy?
- Broadcasting in NumPy is a powerful mechanism that allows arithmetic operations between arrays of different shapes by automatically expanding the smaller array to match the shape of the larger one without making unnecessary copies of data. This enables efficient element-wise operations and vectorized computations.

 - **How Broadcasting Works**
 - When performing operations on two arrays, NumPy compares their shapes starting from the trailing (rightmost) dimensions and works leftwards. Two dimensions are compatible if:

   - They are equal, or
   - One of them is 1

  - If these conditions are met for all dimensions, NumPy "broadcasts" the smaller array by virtually replicating its data along the dimension(s) where it has size 1, so that it matches the shape of the larger array. If the arrays are incompatible, NumPy raises a ValueError.

  - If the arrays have different numbers of dimensions, the shape of the smaller array is padded on the left with ones until both shapes have the same length.



```python
# Example 1: Broadcasting a scalar with a 1D array
import numpy as np

a = np.array([1.0, 2.0, 3.0])
b = 2.0  # scalar

result = a * b
print(result)

# Here, the scalar b is conceptually stretched to match the shape of a (which is (3,)), and the multiplication is done element-wise.
# NumPy does this efficiently without actually copying the scalar multiple times
```

```python
# Example 2: Broadcasting a 1D array with a 2D array

import numpy as np

a1 = np.array([2, 4, 6])          # shape (3,)
a2 = np.array([[1, 3, 5],        # shape (2, 3)
               [7, 9, 11]])

result = a1 + a2
print(result)

# In this example, a1 has shape (3,) and a2 has shape (2, 3).
# NumPy pads a1â€™s shape to (1, 3) and then broadcasts it along the first dimension to match (2, 3). The addition is then performed element-wise
```

4. Explain the use of the groupby() method in Pandas?
- The groupby() method in Pandas is a powerful tool for data analysis that allows you to split a DataFrame into groups based on one or more columns. This enables efficient aggregation, transformation, and filtering of data within each group. The method follows a "split-apply-combine" strategy:
  - **Splitting**: Data is divided into groups based on unique values in specified column(s).

  - **Applying**: Functions (such as sum, mean, count, etc.) are applied independently to each group.

  - **Combining**: The results are combined into a new DataFrame or Series.

```python
import pandas as pd

data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Clothing', 'Books'],
        'Region': ['North', 'South', 'East', 'West', 'North', 'East'],
        'Sales': [100, 50, 120, 30, 60, 40]}
df = pd.DataFrame(data)
print(df)
```

```python
# Now, let's say you want to find the total sales for each Category. You would use groupby() like this:
category_sales = df.groupby('Category')['Sales'].sum()
print(category_sales)
```

```python
# You can also group by multiple columns. For instance, to find the total sales for each Category within each Region:
category_region_sales = df.groupby(['Category', 'Region'])['Sales'].sum()
print(category_region_sales)
```

15. How does NumPy handle multidimensional arrays?
- NumPy handles multidimensional arrays using its core data structure called the ndarray (N-dimensional array), which can represent arrays of any number of dimensions
 - **Creation**: Multidimensional arrays are created by passing nested lists (or arrays) to the np.array() function. For example, passing a list of lists creates a 2D array, and further nesting increases the number of dimensions.

 - **Shape and Dimensions**: Each array has a .shape attribute (a tuple indicating the size along each dimension) and a .ndim attribute (the number of dimensions). For example, a 2x3 array has shape = (2, 3) and ndim = 2.

 - **Indexing and Slicing**: Elements can be accessed and modified using multi-index notation (e.g., arr[1][2] for the element in the second row, third column) and slices can be taken along any axis (e.g., arr[:, 0] for all elements in the first column).

  - **Reshaping**: Arrays can be reshaped to different dimensions using the reshape() method, as long as the total number of elements remains the same.

  - **Mathematical Operations**: Most mathematical operations (addition, multiplication, broadcasting, etc.) are applied element-wise and work seamlessly across all dimensions, enabling efficient computation on large, multi-dimensional datasets.

17. Explain the difference between apply() and map() in Pandas.
- The key differences between apply() and map() in Pandas are as follows:

- **Scope of Use**:
  - **map**() is used only with Pandas Series (one-dimensional data). It applies a function, dictionary, or Series to each element in the Series independently, making it ideal for simple, element-wise transformations or substitutions.

  - **apply**() is more versatile and can be used with both Series and DataFrames. When used on a Series, it applies a function to each element (similar to map()), but it can handle more complex functions. When used on a DataFrame, apply() can apply a function along an entire row or column, allowing for more sophisticated operations
- **Functionality**:
  - **map**() is best for straightforward value mapping or transformations, such as converting categorical values to numerical codes or substituting values using a dictionary.

 - **apply**() is preferred when you need to perform operations that involve entire rows or columns, such as aggregations, custom calculations, or applying functions that require access to multiple values at once.

21. Why is NumPyâ€™s array slicing faster than Pythonâ€™s list slicing?
- NumPyâ€™s array slicing is faster than Pythonâ€™s list slicing for several key reasons:

 - **Contiguous Memory Storage**: NumPy arrays are stored in a continuous block of memory, which allows for efficient access and manipulation of data. This memory layout enables rapid slicing and data retrieval compared to Python lists, which are arrays of pointers to objects scattered throughout memory

  - **Views Instead of Copies**: When you slice a NumPy array, it typically returns a view of the original data rather than creating a copy. This means slicing is a lightweight operation that avoids unnecessary data duplication, leading to significant speed and memory efficiency gains

  - **Low-Level Optimization**: NumPy is implemented in C and leverages optimized, low-level routines for array operations, including slicing. This allows it to execute slicing operations much faster than Pythonâ€™s high-level list slicing, which involves more overhead
  - **Homogeneous Data Types**: NumPy arrays contain elements of the same data type, enabling efficient processing and memory usage. In contrast, Python lists can contain mixed types, requiring additional checks and slowing down operations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.arange(8)
y1 = 4 * x + 5
y2 = 3 * x + 5
y3 = 2 * x + 5
y4 = x + 5

# Plot each line with a label
plt.plot(x, y1, label="4x+5")
plt.plot(x, y2, label="3x+5")
plt.plot(x, y3, label="2x+5")
plt.plot(x, y4, label="x+5")

# Add title and axis labels
plt.title("Plot Multiple Lines in Matplotlib", fontsize=15)
plt.xlabel("X", fontsize=13)
plt.ylabel("Y", fontsize=13)

# Add a legend to distinguish lines
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 32, 45, 28],
    'Score': [85, 90, 78, 88]
}
df = pd.DataFrame(data)

# Set the threshold
threshold = 30

# Filter rows where 'Age' is greater than the threshold
filtered_df = df[df['Age'] > threshold]

print(filtered_df)
```

10. Create a histogram using Seaborn to visualize a distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data (for demonstration)
data = np.random.randn(1000)

# Create a histogram
sns.histplot(data)

# Add title and labels
plt.title('Distribution Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

11. Perform matrix multiplication using NumPy

```python
import numpy as np

# Define two matrices (2D arrays)
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication using np.dot() or @ operator
result = np.dot(A, B)
# Alternatively: result = A @ B

# Print the result
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nMatrix Multiplication (A * B):")
print(result)
```

13. Create a 3D scatter plot using Plotly.


```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate sample data
df = pd.DataFrame({
    'x': np.random.randn(100),
    'y': np.random.randn(100),
    'z': np.random.randn(100),
    'color': np.random.randn(100)  # Optional, for coloring points
})

# Create a 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z', color='color', title='3D Scatter Plot Example')

# Show the plot
fig.show()
```
